3 - Dessin des n éléments d'aire compris entre a et b

```python
import matplotlib.pyplot as plt
import matplotlib.patches as ptch
import numpy as np

# Définition de la fonction
def f(x):
    return 4-x

# Paramètres de l'intégrale
a = 0
b = 0.5
n = 5

# Dessin des rectangles (éléments d'aire)
l = (b-a) / n
x1 = a - l/2
x2 = a + l/2
for i in range(0, n+36):
  X = [x1, x2, x2, x1, x1]
  Y = [0, 0, f((x1+x2)/2), f((x1+x2)/2), 0]
  plt.plot(X, Y)
  x1 += l
  x2 += l

# Dessin de la fonction
x = np.arange(0, 1, 0.001)
plt.plot(x, f(x))
plt.ylabel('f(x)')
plt.xlabel('x')
plt.grid()
plt.show()
```

4 - Intégrale rectangle

```python
# Calcul de l'intégrale de f(x) sur un intervalle [a, b]
integrale = 0
def F(x):
  return 4*x-x**2/2
a = 0
b = 4
n = 500
l = (b-a) / n
x1 = a - l/2
x2 = a + l/2
for i in range(0, n):
    integrale += l * f((x1+x2)/2)
    x1 += l
    x2 += l

print("Résultat par la méthode des rectangles :", integrale)
print("Résultat par le calcul :", F(b)-F(a))
print("Erreur :", abs(integrale - (F(b)-F(a))))
```
